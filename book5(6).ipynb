```python
expr = marks.copy()
```

```python
#access by last digit of registeration no
def student(last_digit):
    regno = 782330220000 + last_digit
    return expr.loc[expr['rn'] == regno]

```

```python
def marksheet(last_digit):
    temp = student(last_digit)
    temp = temp.iloc[:,4:]
    temp = temp.dropna(axis = 1)
    return temp
```

```python
def conv_gp(num):
    if (num == 25):
        temp = 0
    if (num >25):
        temp = (num/75)*100
    else:
        temp = (num/25)*100
        
    if (temp >= 90):
        gp = 4.0
        return gp
    if(temp<90 and temp>=80):
        gp = 3.6
        return gp
    if(temp<80 and temp>=70):
        gp = 3.2
        return gp
    if(temp<70 and temp >=60):
        gp = 2.8
        return gp
    if(temp<60 and temp >=50):
        gp = 2.4
        return gp
    if(temp<50 and temp >= 40):
        gp = 2.0
        return gp
    if(temp<40 and temp >=35):
        gp = 1.6
        return gp
    else:
        gp = 0
        return gp
        
```
